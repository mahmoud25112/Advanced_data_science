# **Iris dataset Analysis with notes**


```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
print('Description')
print(iris.DESCR)
print()
```

```python
print(iris.keys())
data = iris.data
target = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
print()
print('Feature Names:', feature_names)
print()
print('Target Names:', target_names)
print()
print('Data Shape:', data.shape)
```

```python
df = pd.DataFrame(data, columns=feature_names)
df['target'] = target
df['target_names'] = target_names[target]
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df, hue='target_names')
plt.show()
```

```python
sns.boxplot(data=df, x='target_names', y='sepal width (cm)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
print('X_train Shape:', X_train.shape)
print('X_test Shape:', X_test.shape)
print('y_train Shape:', y_train.shape)
print('y_test Shape:', y_test.shape)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier


classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
# prompt: now I need to visulzie a confusion matrix with the results from y_pred and also plot a fetaure importancce graph

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
importances = classifier.feature_importances_

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
print()
# Feature Importance
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance_df)
plt.title('Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python

import matplotlib.pyplot as plt

from sklearn import datasets, svm
from sklearn.inspection import DecisionBoundaryDisplay

# import some data to play with
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000),
    svm.SVC(kernel="rbf", gamma=0.7, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = (
    "SVC with linear kernel",
    "LinearSVC (linear kernel)",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 3) kernel",
)

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]

for clf, title, ax in zip(models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        response_method="predict",
        cmap=plt.cm.coolwarm,
        alpha=0.8,
        ax=ax,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors="k")
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # enables 3D plotting
from sklearn import datasets
from sklearn.svm import SVC

# Load the iris dataset and select the first three features
iris = datasets.load_iris()
X = iris.data[:, :3]  # Use three features for 3D visualization
y = iris.target

# For a clear decision hyperplane, we select only two classes (binary classification)
binary_mask = y < 2  # this keeps classes 0 and 1
X = X[binary_mask]
y = y[binary_mask]

# Train a linear SVM
clf = SVC(kernel='linear')
clf.fit(X, y)

# Retrieve the coefficients of the hyperplane: w0*x + w1*y + w2*z + b = 0
w = clf.coef_[0]
b = clf.intercept_[0]

# Create a meshgrid for plotting the decision plane.
# We choose feature 0 for the x-axis and feature 1 for the y-axis.
x_range = np.linspace(X[:, 0].min() - 0.5, X[:, 0].max() + 0.5, 10)
y_range = np.linspace(X[:, 1].min() - 0.5, X[:, 1].max() + 0.5, 10)
xx, yy = np.meshgrid(x_range, y_range)

# Solve for z in the hyperplane equation: w0*x + w1*y + w2*z + b = 0  ->  z = -(w0*x + w1*y + b) / w2
zz = -(w[0] * xx + w[1] * yy + b) / w[2]

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the decision hyperplane
ax.plot_surface(xx, yy, zz, color='yellow', alpha=0.5, rstride=1, cstride=1)

# Scatter plot the training points
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.coolwarm, s=50, edgecolors='k')

# Label the axes with the feature names
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])
plt.title("3D Decision Hyperplane with Linear SVM")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting
from sklearn import datasets
from sklearn.svm import SVC

# Step 1: Load the iris dataset and select the first three features for 3D visualization.
iris = datasets.load_iris()
X = iris.data[:, :3]  # Use the first three features (e.g., sepal length, sepal width, petal length)
y = iris.target

# Step 2: Filter to only two classes to simplify the decision boundary (binary classification).
# Here, we use classes 0 and 1.
binary_mask = y < 2  # This creates a boolean mask where True for classes 0 and 1.
X = X[binary_mask]
y = y[binary_mask]

# Step 3: Train a linear Support Vector Machine (SVM) classifier.
# A linear SVM will give us a flat hyperplane (plane) in 3D.
clf = SVC(kernel='linear')
clf.fit(X, y)

# Step 4: Extract the coefficients (w) and the intercept (b) of the hyperplane.
# The hyperplane equation is: w0*x + w1*y + w2*z + b = 0.
w = clf.coef_[0]  # Coefficients for the three features.
b = clf.intercept_[0]  # Intercept (bias term).

# Step 5: Create a meshgrid for plotting the decision hyperplane.
# We choose ranges for x and y based on the data.
x_range = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 10)
y_range = np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 10)
xx, yy = np.meshgrid(x_range, y_range)

# Compute the corresponding z values for each (x, y) point using the hyperplane equation:
# z = -(w0*x + w1*y + b) / w2
zz = -(w[0] * xx + w[1] * yy + b) / w[2]

# Step 6: Create a 3D plot.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the decision hyperplane as a semi-transparent surface.
ax.plot_surface(xx, yy, zz, color='yellow', alpha=0.5, rstride=1, cstride=1)

# Scatter plot the data points to show the actual classes.
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.coolwarm, s=50, edgecolors='k')

# Label the axes using the iris dataset's feature names.
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])

# Step 7: Adjust the viewing angle.
# Use the view_init method with elev (elevation) and azim (azimuth) parameters.
# Elevation (30 degrees) sets the height angle above the xy-plane.
# Azimuth (60 degrees) rotates the view around the z-axis.
ax.view_init(elev=30, azim=60)

# Optional: Set a title for the plot.
plt.title("3D Decision Hyperplane Visualization from Different Angle")

# Step 8: Display the plot.
plt.show()
```
